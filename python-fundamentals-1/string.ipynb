```python
# count the frequuency of a particular character in a provided string.
# eg 'how are you' is the string, the frequency of h in the string is 2
s= input('Enter the string : ')
term = input('Enter the term: ')

count = 0
for i in s:
    if i==term:
        count+=1

print(count)
        
```

```python
# write a program which can remove a particular character from a string
s = input('Enter the string: ')
pr = input('Enter the particular char: ')

result = ''
for i in s:
    if i != pr:
        result = result+i

print(result)
```

```python
# Wtrite a program that can check wheter a given string is palindrome or not.
# nitin
# abba
s = input('Enter the string: ')
flag = True
for i in range(0,len(s)//2):
    if s[i] != s[len(s)-i-1]:
        flag = False
        print('Not a palindrome')
        break

if flag:
    print('Palindrome')


print(s[len(s)-i-1])

```

```python
# write a program to count the number of wrods in a string without split()
s =  input('Enter the string s: ')
L = []
temp = ''
for i in s:
    if i != ' ':
        temp = temp+i
    else:
        L.append(temp)
        temp = ''

L.append(temp)
print(L)
```

```python
# Write a python program to convert a string to title case without using the title()

s = input('enter the string s: ')
L = []
for i in s.split():
    L.append(i[0].upper() + i[1:].lower())

print(' '.join(L))
```

```python
# write a program that can convert an integer to string

num = int(input('Enter the number: '))
result = ''
digits = '0123456789'

while num!=0:
    result = digits[num%10] + result
    num = num//10

print(result)

type(result)
```
